### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
df_minmax=df.copy()
age_mm=df_minmax['Age'].interpolate()
age_min=age_mm.min()
age_max=age_mm.max()
df_minmax=(age_mm-age_min)/(age_max-age_min)
df_minmax
```

```python
data_ds = df.copy()
age_ds= data_ds['Age']
max_age = age_ds.abs().max()
temp = len(str(int(max_age)))
data_ds['DecimalScalAge']=age_ds/(10**temp)
data_ds[['Age','DecimalScalAge']]
```

```python
x=df['Age']
mean=df['Age'].mean()
standard_deviation = df['Age'].std()
x_scaled=(x-mean)/standard_deviation
data_ds['ZScoreAge']=x_scaled
data_ds['MinMaxAge']=df_minmax
data_ds[['Age','DecimalScalAge','ZScoreAge','MinMaxAge']]
```
